```
import tensorflow as tf
import numpy as np
print("TF Version", tf.__version__)
```

```
sess = tf.InteractiveSession()
```

```
X, Y = np.mgrid[-4.0:4:0.01, -4.0:4:0.01]
```

```
xs = tf.constant(X.astype(np.float32))
ys = tf.constant(Y.astype(np.float32))
```

```
tf.global_variables_initializer().run() #init variables
```

```
#Compute Gaussian
zs = tf.exp(-(xs**2+ys**2)/2.0)
```

```
#plot Gaussian
import matplotlib.pyplot as plt
plt.imshow(zs.eval())
plt.tight_layout()
plt.show()
```

```
#Compute Sine and Cosine
zsc = tf.math.cos(xs,name=None)
zss = tf.math.sin(ys,name=None)
```

```
#plot Sine
plt.imshow(zss.eval())
plt.tight_layout()
plt.show()
```

```
#plot Cosine
plt.imshow(zsc.eval())
plt.tight_layout()
plt.show()
```

```
#Compute product of Gaussian and Sine
zm = zs*zss
#plot the product zm
plt.imshow(zm.eval())
plt.tight_layout()
plt.show()
```
